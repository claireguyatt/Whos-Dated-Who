```python
# set up
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# get data (collected from previously ran scripts)
wdw_df = pd.read_csv('data/wdw_data.csv')
print(wdw_df.head(), "\n")
print(wdw_df.info())
```

```python
# explore data

wdw_df['num_relations'].value_counts()[:20].plot(kind='barh', title="Most to least common number of relationships")
```

```python
# The most common number of relationships is 0. Doesn't make sense - to be included in the dataset each person
# had to have at least one relationship. Remove the incorrect data (all rows with num_relations==0)

wdw_df = wdw_df[wdw_df.num_relations != 0]
```

```python
# explore null values

sns.heatmap(wdw_df.isnull(), cbar=False)

print("Percent missing per feature: \n\n", wdw_df.isnull().sum()/len(wdw_df)*100)
```

```python
# Religion feature has ~60% missing values --> drop Religion column
wdw_df = wdw_df.drop(['Religion'], axis=1)
```
